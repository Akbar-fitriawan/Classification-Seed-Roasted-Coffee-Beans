# **Load Model**

```python
def load_model(model_path):
    """Load the pre-trained model from the given path."""
    model = tf.keras.models.load_model(model_path)
    return model
```

```python

def prediction(file, model):
    """Predict the class of the given image using the loaded model."""
    # Load and preprocess the image
    img = tf.keras.utils.load_img(file, target_size=(224, 224))
    img_array = tf.keras.utils.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Predict the class of the image
    predictions = model.predict(img_array, batch_size=32)
    idx = np.argmax(predictions)
    labels = ['Dark', 'Green', 'Light', 'Medium']

    # Display the prediction
    print('Prediction is a {}'.format(labels[idx]))
    plt.imshow(tf.keras.utils.load_img(file))
    plt.title(f"Prediction: {labels[idx]}")
    plt.axis(False)
    plt.show()

    # Print probabilities
    print('Prediction probabilities:')
    for i, label in enumerate(labels):
        print(f"{label}: {predictions[0][i]:.4f}")
```
